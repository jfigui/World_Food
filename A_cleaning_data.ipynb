## Useful functions

```python
# Function to find unique strings in a column
def find_unique_strings(x):
    strings = []
    for entry in x.unique():
        entry.split(',')
        strings.extend(entry.split(','))
    strings = np.array(strings)
    return np.unique(strings)
```

## Correct states column

```python
# Transform states data
def  transform_states(s):
    # correct entries with tags    
    s=s.str.replace('-', ' ')
    
    return(s)
```

```python
# Transform states data
def  add_state(s):
    items_incomplete = ['Brands', 'Categories', 'Packaging', 'Product name', 'Quantity']
    
    for it in items_incomplete:
        s.loc[~s.str.contains(it)] = it+' completed,'+s[~s.str.contains(it)]        
    return(s)
```

## Change code in duplicated codes

```python
code = df.code
```

```python
# convert code to int. It cannot handle the conversion directly so we first convert to float
code = code.astype(float)
code = code.astype(int)
code.dtype
```

```python
# Check if all entries have a unique code
ncode_unique = code.unique().size
ncode_unique
```

```python
# number of entries with duplicated code
nrows = code.size
nrows-ncode_unique
```

```python
# find codes that are duplicated
duplicated_codes = code[code.duplicated()]
```

```python
code[code == duplicated_codes.iloc[0]].index
```

```python
# Assign a new code to the duplicated codes
for dc in duplicated_codes:
    indices = code[code == dc].index
    for i, ind in enumerate(indices):            
        code.loc[ind] = code.loc[ind]+i
```

```python
# Check that no codes are duplicated now
duplicated_codes = code[code.duplicated()]
duplicated_codes
```

```python
df['code'] = code
df['code'].dtype
```
